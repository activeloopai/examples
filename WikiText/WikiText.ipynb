```python
#Loading Packages
from datasets import load_dataset
import deeplake
import numpy as np
from tqdm import tqdm
```

```python
#Load Dataset
dataset = load_dataset("wikitext",'wikitext-2-raw-v1')
```

```python
# Initialize Deeplake 
ds_train = deeplake.empty('./wikitext_deeplake',overwrite=True)
```

```python
dataset
```

```python
# dataset['train']['text']
```

```python
# Creating Tensors
with ds_train:
    ds_train.create_tensor("Paragraph", htype="text")
    ds_train.create_tensor("tokens",htype='text')  
```

```python
# Checking Lengths of Train, test and valadation data
len(dataset['train']),len(dataset['test']),len(dataset['validation'])
```

```python
from tqdm import tqdm
with ds_train:
    for i in tqdm(range(1,len(dataset['train']))):
        if len(dataset['train']['text'][i]) < 1:
            continue
        if dataset['train']['text'][i][1] == '=':
            ds_train['tokens'].append(dataset['train']['text'][i])    
        if dataset['train']['text'][i][1] != '=':
            ds_train['Paragraph'].append(dataset['train']['text'][i]) 
```
